The goal of this MVP is to build a first version of a scenario conditioning engine.

Given a set of user-defined stressed factors, the engine should estimate coherent implied shocks for the remaining factors in the universe.

For example, if the user pins:

- S&P 500 futures: -20%
- 10Y Treasury yield: +150 bps

the engine should return implied shocks for related equity, rates, commodity, FX, and metal factors.

## 8. Multicollinearity Check
Condition number, high correlation pins, pseudo-inverse / ridge fallback

thoughts:
first check with the pinned factors

one example (2 factors):
- if highly correlated factors have opposite stress (ex: sp500 +10, nasdaq -10), flag inconsistency
- if highly correlated factors have the same-direction stress (ex: sp500 +10, nasdaq +12), allow it but monitor redundancy / magnitude

important regime note:
- cross-asset relationships such as equity vs treasury futures are not hard-coded as always opposite
- stock and bond futures can move together in inflation, liquidity, or policy-shock regimes
- therefore equity-rate conditional sign conflicts should usually be warning / regime-review, not automatic hard inconsistency


```python
def status_helper_func(num_direction_conflict, num_magnitude_warning):
    """
    Decide overall status based on direction conflicts and magnitude warnings.

    """

    if num_direction_conflict > 0:
        status = "review_required"
        main_issue = "direction_conflict"
        message = (
            "Some highly correlated pinned shocks move in mutually inconsistent "
            "directions. This scenario should be reviewed."
        )

    elif num_magnitude_warning > 0:
        status = "warning"
        main_issue = "large_magnitude_deviation"
        message = (
            "No hard high-correlation direction conflict was found, but some pinned "
            "shocks have notable regime-dependent sign or magnitude deviations."
        )

    else:
        status = "ok"
        main_issue = "none"
        message = "Pinned shocks look directionally consistent with the model."

    return status, main_issue, message


def check_pinned_multicollinearity(
    shocks,
    cov,
    mean=None,
    magnitude_warning_z=4.0,
    ridge=1e-8,
    min_implied_abs=1e-6,
    strong_corr_threshold=0.75,
    soft_direction_warning_z=2.0
):
    """
    Model-based pinned factor consistency check.

    Main logic:
    1. For each pinned factor, use the other pinned factors to compute
       the model-implied conditional value.
    2. If a highly correlated pinned pair has mutually inconsistent signs,
       flag review_required.
    3. If a cross-asset / weaker-correlation conditional sign conflict appears,
       downgrade it to warning because regimes can switch.
    4. If direction is consistent but magnitude is very far away,
       flag warning.
    5. Otherwise, return ok.

    This keeps the check focused on true multicollinearity problems,
    instead of hard-coding cross-asset assumptions like stocks and bonds
    always moving in opposite directions.
    """

    # Convert shocks to pandas Series
    if not isinstance(shocks, pd.Series):
        shocks = pd.Series(shocks)

    pinned = list(shocks.index)

    # Convert mean to pandas Series
    # If mean is not provided, assume zero mean for pinned factors
    if mean is None:
        mean = pd.Series(0.0, index=pinned)
    elif not isinstance(mean, pd.Series):
        mean = pd.Series(mean)

    mean = mean.reindex(pinned)

    num_direction_conflict = 0
    num_magnitude_warning = 0

    max_abs_z_error = 0.0
    worst_factor = None
    direction_conflict_factors = []
    magnitude_warning_factors = []
    soft_direction_conflict_factors = []

    # Loop through each pinned factor as the target
    for target in pinned:

        # Use the remaining pinned factors to explain this target
        others = [f for f in pinned if f != target]

        actual = shocks.loc[target]

        mu_target = mean.loc[target]
        mu_others = mean.loc[others].to_numpy()

        # Covariance blocks
        Sigma_iO = cov.loc[target, others].to_numpy().reshape(1, -1)
        Sigma_OO = cov.loc[others, others].to_numpy()
        Sigma_ii = cov.loc[target, target]

        # Add ridge for numerical stability
        Sigma_OO_ridge = Sigma_OO + ridge * np.eye(len(others))

        x_others = shocks.loc[others].to_numpy()

        # Model-implied conditional value
        implied = (
            mu_target
            + Sigma_iO
            @ np.linalg.pinv(Sigma_OO_ridge)
            @ (x_others - mu_others)
        ).item()

        # Conditional variance
        cond_var = (
            Sigma_ii
            - Sigma_iO
            @ np.linalg.pinv(Sigma_OO_ridge)
            @ Sigma_iO.T
        ).item()

        cond_std = np.sqrt(max(cond_var, 1e-12))

        # Standardized magnitude difference
        z_error = (actual - implied) / cond_std
        abs_z_error = abs(z_error)

        if abs_z_error > max_abs_z_error:
            max_abs_z_error = abs_z_error
            worst_factor = target

        # Hard direction review is reserved for true high-correlation conflicts.
        # Cross-asset sign conflicts can be regime-dependent, so they become warnings.
        strong_pair_conflict = False
        for other in others:
            other_actual = shocks.loc[other]
            if abs(actual) < min_implied_abs or abs(other_actual) < min_implied_abs:
                continue

            other_var = cov.loc[other, other]
            corr_denom = np.sqrt(max(Sigma_ii, 1e-12) * max(other_var, 1e-12))
            corr = cov.loc[target, other] / corr_denom

            if abs(corr) >= strong_corr_threshold:
                expected_sign = np.sign(corr) * np.sign(other_actual)
                if expected_sign != 0 and np.sign(actual) != expected_sign:
                    strong_pair_conflict = True
                    break

        # Main review rule:
        # Only check direction if implied value is large enough to be meaningful.
        # This avoids overreacting when model implied shock is basically zero.
        if abs(implied) >= min_implied_abs:
            if np.sign(actual) != np.sign(implied):
                if strong_pair_conflict:
                    num_direction_conflict += 1
                    direction_conflict_factors.append(target)
                else:
                    if abs_z_error >= soft_direction_warning_z:
                        num_magnitude_warning += 1
                        soft_direction_conflict_factors.append(target)

            # Direction is correct, but magnitude is very far away
            elif abs_z_error >= magnitude_warning_z:
                num_magnitude_warning += 1
                magnitude_warning_factors.append(target)

        else:
            # If implied value is close to zero, direction is not meaningful.
            # Only give warning if actual shock is extremely far from implied.
            if abs_z_error >= magnitude_warning_z:
                num_magnitude_warning += 1
                magnitude_warning_factors.append(target)

    status, main_issue, message = status_helper_func(
        num_direction_conflict,
        num_magnitude_warning
    )

    return {
        "status": status,
        "main_issue": main_issue,
        "message": message,
        "details": {
            "num_pinned_factors": len(pinned),
            "num_direction_conflicts": num_direction_conflict,
            "num_magnitude_warnings": num_magnitude_warning,
            "max_abs_z_error": max_abs_z_error,
            "worst_factor": worst_factor,
            "direction_conflict_factors": direction_conflict_factors,
            "magnitude_warning_factors": magnitude_warning_factors,
            "soft_direction_conflict_factors": soft_direction_conflict_factors,
            "strong_corr_threshold": strong_corr_threshold,
            "soft_direction_warning_z": soft_direction_warning_z
        }
    }
```

## 10. Baseline Method: Multi Conditional Gaussian

### Step 1: Compute Returns

We first convert raw price data into percentage returns.

$$
R_t = \frac{P_t - P_{t-1}}{P_{t-1}}
$$

This step transforms the price matrix into a return matrix. Each column represents one market factor, and each row represents one historical date.
    return price_df.pct_change().dropna()

### Step 2: Compute Historical Mean

After computing returns, we calculate the historical average return for each factor.

$$
\mu = E[R]
$$

The mean vector represents the normal historical return level of each factor.

### Step 3: Compute Historical Covariance Matrix

Next, we calculate the covariance matrix across all factor returns.

$$
\Sigma = Cov(R)
$$

The covariance matrix captures how different market factors historically move together. This is the key input for the conditional Gaussian model.

### Step 4: Split Given and Target Factors

We separate the factors into two groups:

- **Given factors**: factors pinned by the user
- **Target factors**: factors whose implied shocks we want to estimate

Let:

$$
A = \text{given factors}
$$

$$
B = \text{target factors}
$$

We then split the mean vector and covariance matrix into:

$$
\mu_A = \text{mean of given factors}
$$

$$
\mu_B = \text{mean of target factors}
$$

$$
\Sigma_{AA} = \text{covariance among given factors}
$$

$$
\Sigma_{BA} = \text{covariance between target factors and given factors}
$$

This prepares the correct covariance blocks for the conditional mean formula.

### Step 5: Compute Conditional Mean

The conditional Gaussian formula is:

$$
E[B \mid A = a]
=
\mu_B + \Sigma_{BA}\Sigma_{AA}^{-1}(a - \mu_A)
$$

Where:

- \(A\) is the set of given factors
- \(B\) is the set of target factors
- \(a\) is the user-defined stress scenario
- \(\mu_A\) is the historical mean of the given factors
- \(\mu_B\) is the historical mean of the target factors
- \(\Sigma_{AA}\) is the covariance matrix among given factors
- \(\Sigma_{BA}\) is the covariance between target factors and given factors

The intuition is:

$$
\text{Conditional Mean}
=
\text{Normal Mean}
+
\text{Covariance-Based Adjustment}
$$

The model adjusts the expected target factor shocks based on how far the user-pinned scenario is from the historical average.


```python
def compute_returns(price_df):
    return price_df.pct_change().dropna()

def compute_mean(returns_df):
    return returns_df.mean()

#sample covariance
# def compute_covariance(returns_df):
#     return returns_df.cov()

# EWMA covariance
def compute_ewma_covariance(returns_df, lambda_=0.9):
    """
    EWMA covariance: recent dates get higher weights.
    """

    demeaned = returns_df - returns_df.mean()
    n = len(demeaned)

    # Oldest date gets smallest weight, newest date gets largest weight
    weights = np.array([
        lambda_ ** age
        for age in range(n - 1, -1, -1)
    ])

    # Normalize so all weights sum to 1
    weights = weights / weights.sum()

    weighted_returns = demeaned.multiply(weights, axis=0)

    ewma_cov = weighted_returns.T.dot(demeaned)

    return ewma_cov

def compute_covariance(returns_df, method="sample", lambda_=0.9):
    if method == "sample":
        return returns_df.cov()

    if method == "ewma":
        return compute_ewma_covariance(returns_df, lambda_=lambda_)

    raise ValueError("method must be either 'sample' or 'ewma'")

# split the mean and covariance for conditional mean calculation
def split_for_conditional_mean(mu, cov, given_factors, target_factors):
    mu_A = mu.loc[given_factors]
    mu_B = mu.loc[target_factors]

    Sigma_AA = cov.loc[given_factors, given_factors]
    Sigma_BA = cov.loc[target_factors, given_factors]

    return mu_A, mu_B, Sigma_AA, Sigma_BA

# compute the conditional mean of target factors given the values of given factors
def compute_conditional_mean(mu_A, mu_B, Sigma_AA, Sigma_BA, scenario):
    if isinstance(scenario, pd.Series):
        scenario = scenario.reindex(mu_A.index)
    else:
        scenario = pd.Series(scenario, index=mu_A.index)

    delta = scenario - mu_A
    adjustment = Sigma_BA.to_numpy() @ np.linalg.inv(Sigma_AA.to_numpy()) @ delta.to_numpy()
    conditional_mean = mu_B + pd.Series(adjustment, index=mu_B.index)
    return conditional_mean

# main function to compute the conditional mean of target factors given the values of given factors
def conditional_mean_engine(returns_df, given_factors, target_factors, scenario, covariance_method="sample", ewma_lambda=0.9):

    # 1. Compute model inputs
    mu = compute_mean(returns_df)

    # cov = compute_covariance(returns_df)
    cov = compute_covariance(
        returns_df,
        method=covariance_method,
        lambda_=ewma_lambda
    )

    # 2. Check pinned/given factor consistency before conditioning
    pin_check = check_pinned_multicollinearity(
        shocks=scenario,
        cov=cov,
        mean=mu
    )

    pin_check["details"]["covariance_method"] = covariance_method
    if covariance_method == "ewma":
        pin_check["details"]["ewma_lambda"] = ewma_lambda

    # 3. Split matrix for conditional mean
    mu_A, mu_B, Sigma_AA, Sigma_BA = split_for_conditional_mean(
        mu, cov, given_factors, target_factors
    )

    # 4. Compute conditional mean
    conditional_mean = compute_conditional_mean(
        mu_A, mu_B, Sigma_AA, Sigma_BA, scenario
    )

    # 5. Return both result and check
    return {
        "conditional_mean": conditional_mean,
        "pin_check": pin_check
    }
```

```python
given_factors = ['sp500_future', 'nasdaq100_future', 'us10y_note_future']  # remove extra quotes here
target_factors = [
    factor for factor in returns_df.columns
    if factor not in given_factors
]

# pretend the sp500 decrease 20%，bond increase 0.87%
scenario = pd.Series(
    {
        "sp500_future": -0.10,
        "nasdaq100_future": -0.12,
        "us10y_note_future": 0.0087
    }
)

result = conditional_mean_engine(
    returns_df,
    given_factors,
    target_factors,
    scenario,
    covariance_method="ewma",
    ewma_lambda=0.9
)

scenario_result = result["conditional_mean"].to_frame("implied_return")
scenario_result["implied_return_pct"] = scenario_result["implied_return"] * 100

print(scenario_result)
result["pin_check"] #one problem, wti goes down when index goes down, seems not correct????
```

```python
def show_ewma_weights(returns_df, lambda_=0.94, last_n=20):
    n = len(returns_df)

    weights = np.array([
        lambda_ ** age
        for age in range(n - 1, -1, -1)
    ])
    weights = weights / weights.sum()

    weight_df = pd.DataFrame({
        "date": returns_df.index,
        "weight": weights,
        "weight_pct": weights * 100
    })

    return weight_df.tail(last_n)
```

```python
import html

import numpy as np
import pandas as pd
import ipywidgets as widgets
from IPython.display import HTML, clear_output, display

required_objects = ["returns_df", "conditional_mean_engine"]
missing_objects = [name for name in required_objects if name not in globals()]
if missing_objects:
    raise RuntimeError(
        "Run the previous data preparation and engine cells first. "
        f"Missing: {', '.join(missing_objects)}"
    )

factor_universe = list(returns_df.columns)
if not factor_universe:
    raise ValueError("returns_df has no factor columns to pin.")

ticker_lookup = globals().get("tickers", {})


def _ticker_for_factor(factor):
    return ticker_lookup.get(factor, factor)


def _label_for_factor(factor):
    ticker = _ticker_for_factor(factor)
    return str(factor) if ticker == factor else f"{factor} | {ticker}"


factor_labels = [_label_for_factor(factor) for factor in factor_universe]
label_to_factor = dict(zip(factor_labels, factor_universe))
pinned_shock_widgets = {}

status_styles = {
    "ok": {"label": "OK", "border": "#2f9e5b", "background": "#eaf7ef", "text": "#146c36"},
    "warning": {"label": "Warning", "border": "#d99021", "background": "#fff6df", "text": "#8a5a00"},
    "review_required": {"label": "Review Required", "border": "#d33f49", "background": "#fdebec", "text": "#9e1f2b"},
}

search_box = widgets.Text(
    value="",
    placeholder="Search factor or ticker",
    description="Search",
    layout=widgets.Layout(width="100%"),
)

factor_select = widgets.Select(
    description="Tickers",
    rows=min(max(len(factor_labels) + 1, 8), 18),
    layout=widgets.Layout(width="100%", height="320px"),
)

pinned_box = widgets.VBox(
    layout=widgets.Layout(
        width="100%",
        border="1px solid #d9dee7",
        padding="10px",
        min_height="120px",
    )
)

result_output = widgets.Output(
    layout=widgets.Layout(
        width="100%",
        border="1px solid #d9dee7",
        padding="12px",
        margin="12px 0 0 0",
    )
)

clear_button = widgets.Button(
    description="Clear Pins",
    icon="times",
    button_style="",
    layout=widgets.Layout(width="120px"),
)

run_button = widgets.Button(
    description="Run",
    icon="play",
    button_style="primary",
    layout=widgets.Layout(width="90px"),
)

sample_checkbox = widgets.Checkbox(
    value=True,
    description="Sample",
    indent=False,
    layout=widgets.Layout(width="90px"),
)

ewma_checkbox = widgets.Checkbox(
    value=False,
    description="EWMA",
    indent=False,
    layout=widgets.Layout(width="90px"),
)

dashboard_ewma_lambda = 0.94


def _filtered_factor_options(query):
    query = query.strip().lower()
    options = []
    for label in factor_labels:
        factor = label_to_factor[label]
        ticker = _ticker_for_factor(factor)
        searchable_text = f"{factor} {ticker}".lower()
        if not query or query in searchable_text:
            options.append((label, label))

    if not options:
        return [("No matching tickers", "")]
    return [("Select a factor to pin", "")] + options


def _scenario_from_widgets():
    return pd.Series(
        {
            factor: shock_widget.value / 100.0
            for factor, shock_widget in pinned_shock_widgets.items()
        },
        dtype=float,
    )


def _selected_covariance_methods():
    methods = []
    if sample_checkbox.value:
        methods.append(("sample", "Sample"))
    if ewma_checkbox.value:
        methods.append(("ewma", f"EWMA lambda={dashboard_ewma_lambda:.2f}"))
    return methods


def _status_card(pin_check, scenario):
    status = pin_check.get("status", "warning")
    style = status_styles.get(status, status_styles["warning"])
    details = pin_check.get("details", {})
    max_abs_z_error = details.get("max_abs_z_error")
    max_abs_z_error_text = "n/a" if max_abs_z_error is None else f"{float(max_abs_z_error):.2f}"
    pinned_summary = ", ".join(
        f"{html.escape(str(factor))}: {shock * 100:+.2f}%"
        for factor, shock in scenario.items()
    )

    return f"""
    <div style="border-left:5px solid {style['border']}; background:{style['background']};
                color:{style['text']}; padding:12px 14px; margin-bottom:12px;">
        <div style="font-size:16px; font-weight:700; margin-bottom:4px;">Status: {style['label']}</div>
        <div style="margin-bottom:6px;">{html.escape(str(pin_check.get('message', '')))}</div>
        <div style="font-size:12px; color:#333;">
            Pinned: {pinned_summary}<br>
            Issue: {html.escape(str(pin_check.get('main_issue', 'n/a')))} |
            Direction conflicts: {details.get('num_direction_conflicts', 'n/a')} |
            Magnitude warnings: {details.get('num_magnitude_warnings', 'n/a')} |
            Max abs z-error: {max_abs_z_error_text} |
            Worst factor: {html.escape(str(details.get('worst_factor', 'n/a')))}
        </div>
    </div>
    """


def _results_table_html(scenario_result):
    rows = []
    for row in scenario_result.itertuples(index=False):
        value = float(row.implied_shock_pct)
        if value > 0:
            color = "#128a3c"
            direction_color = "#128a3c"
        elif value < 0:
            color = "#c7382f"
            direction_color = "#c7382f"
        else:
            color = "#555"
            direction_color = "#555"

        rows.append(
            "<tr>"
            f"<td>{html.escape(str(row.factor))}</td>"
            f"<td>{html.escape(str(row.ticker))}</td>"
            f"<td style='color:{color}; font-weight:700; text-align:right;'>{value:+.3f}%</td>"
            f"<td style='color:{direction_color}; font-weight:700;'>{html.escape(str(row.direction))}</td>"
            "</tr>"
        )

    return """
    <table style="border-collapse:collapse; width:100%; font-size:13px;">
        <thead>
            <tr style="background:#f4f6f8; border-bottom:1px solid #d9dee7;">
                <th style="text-align:left; padding:8px;">factor</th>
                <th style="text-align:left; padding:8px;">ticker</th>
                <th style="text-align:right; padding:8px;">implied_shock_pct</th>
                <th style="text-align:left; padding:8px;">direction</th>
            </tr>
        </thead>
        <tbody>
    """ + "\n".join(
        row.replace("<td>", "<td style='padding:8px; border-bottom:1px solid #edf0f3;'>")
        for row in rows
    ) + """
        </tbody>
    </table>
    """


def _scenario_result_table(conditional_mean):
    scenario_result = conditional_mean.to_frame("implied_return")
    scenario_result.index.name = "factor"
    scenario_result["ticker"] = [_ticker_for_factor(factor) for factor in scenario_result.index]
    scenario_result["implied_shock_pct"] = scenario_result["implied_return"] * 100.0
    scenario_result["direction"] = np.select(
        [
            scenario_result["implied_shock_pct"] > 0,
            scenario_result["implied_shock_pct"] < 0,
        ],
        ["Increase", "Decrease"],
        default="Flat",
    )
    scenario_result = scenario_result.reset_index()[
        ["factor", "ticker", "implied_shock_pct", "direction"]
    ]
    return scenario_result.sort_values(
        "implied_shock_pct",
        key=lambda values: values.abs(),
        ascending=False,
    )


def _method_panel_html(method_label, pin_check, scenario, scenario_result):
    return f"""
    <div style="flex:1 1 480px; min-width:420px; border:1px solid #d9dee7;
                padding:12px; background:#fff; box-sizing:border-box;">
        <div style="font-size:15px; font-weight:700; margin:0 0 10px 0; color:#222;">
            {html.escape(method_label)}
        </div>
        {_status_card(pin_check, scenario)}
        {_results_table_html(scenario_result)}
    </div>
    """


def _render_results(*_):
    with result_output:
        clear_output(wait=True)

        if not pinned_shock_widgets:
            display(
                HTML(
                    "<div style='color:#555;'>Select a factor on the left to pin it. "
                    "Pinned shocks and conditional mean results will appear here.</div>"
                )
            )
            return

        scenario = _scenario_from_widgets()
        given_factors = list(scenario.index)
        target_factors = [factor for factor in factor_universe if factor not in given_factors]

        selected_methods = _selected_covariance_methods()
        if not selected_methods:
            display(HTML("<div style='color:#555;'>Select Sample, EWMA, or both.</div>"))
            return

        panels = []
        for covariance_method, method_label in selected_methods:
            try:
                result = conditional_mean_engine(
                    returns_df=returns_df,
                    given_factors=given_factors,
                    target_factors=target_factors,
                    scenario=scenario,
                    covariance_method=covariance_method,
                    ewma_lambda=dashboard_ewma_lambda,
                )
            except Exception as exc:
                display(
                    HTML(
                        "<div style='border-left:5px solid #d33f49; background:#fdebec; "
                        "color:#9e1f2b; padding:12px 14px;'>"
                        f"Dashboard could not run {html.escape(method_label)}: {html.escape(str(exc))}"
                        "</div>"
                    )
                )
                return

            conditional_mean = result["conditional_mean"]
            if conditional_mean.empty:
                panels.append(
                    f"""
                    <div style="flex:1 1 480px; min-width:420px; border:1px solid #d9dee7;
                                padding:12px; background:#fff; box-sizing:border-box;">
                        <div style="font-size:15px; font-weight:700; margin:0 0 10px 0; color:#222;">
                            {html.escape(method_label)}
                        </div>
                        {_status_card(result["pin_check"], scenario)}
                        <div style='color:#555;'>All factors are pinned, so there are no remaining target factors.</div>
                    </div>
                    """
                )
                continue

            scenario_result = _scenario_result_table(conditional_mean)
            panels.append(
                _method_panel_html(
                    method_label=method_label,
                    pin_check=result["pin_check"],
                    scenario=scenario,
                    scenario_result=scenario_result,
                )
            )

        display(
            HTML(
                "<div style='display:flex; gap:14px; align-items:flex-start; "
                "overflow-x:auto; width:100%;'>"
                + "\n".join(panels)
                + "</div>"
            )
        )


def _refresh_pinned_panel():
    if not pinned_shock_widgets:
        pinned_box.children = (
            widgets.HTML("<div style='color:#555;'>No pinned factors yet.</div>"),
        )
        return

    header = widgets.HBox(
        [
            widgets.HTML("<b style='display:inline-block;width:260px;'>Factor</b>"),
            widgets.HTML("<b>Shock (%)</b>"),
        ]
    )
    rows = [header]

    for factor, shock_widget in pinned_shock_widgets.items():
        remove_button = widgets.Button(
            icon="trash",
            tooltip=f"Remove {factor}",
            layout=widgets.Layout(width="34px"),
        )

        def _remove_pinned(_, factor=factor):
            pinned_shock_widgets.pop(factor, None)
            _refresh_pinned_panel()
            _render_results()

        remove_button.on_click(_remove_pinned)

        factor_label = html.escape(str(factor))
        ticker_label = html.escape(str(_ticker_for_factor(factor)))
        rows.append(
            widgets.HBox(
                [
                    widgets.HTML(
                        f"<div style='width:260px;'><b>{factor_label}</b><br>"
                        f"<span style='color:#666;font-size:12px;'>{ticker_label}</span></div>"
                    ),
                    shock_widget,
                    remove_button,
                ],
                layout=widgets.Layout(
                    align_items="center",
                    margin="4px 0",
                    width="100%",
                ),
            )
        )

    pinned_box.children = tuple(rows)


def _add_factor_to_pins(factor):
    if factor in pinned_shock_widgets:
        return

    shock_input = widgets.FloatText(
        value=0.0,
        step=0.25,
        layout=widgets.Layout(width="110px"),
    )
    shock_input.observe(_render_results, names="value")
    pinned_shock_widgets[factor] = shock_input
    _refresh_pinned_panel()
    _render_results()


def _on_search_change(change):
    factor_select.options = _filtered_factor_options(change["new"])
    factor_select.value = ""


def _on_factor_selected(change):
    selected_label = change["new"]
    if not selected_label:
        return

    factor = label_to_factor.get(selected_label)
    if factor is None:
        return

    _add_factor_to_pins(factor)
    factor_select.value = ""


def _clear_pins(_):
    pinned_shock_widgets.clear()
    _refresh_pinned_panel()
    _render_results()


search_box.observe(_on_search_change, names="value")
factor_select.observe(_on_factor_selected, names="value")
sample_checkbox.observe(_render_results, names="value")
ewma_checkbox.observe(_render_results, names="value")
clear_button.on_click(_clear_pins)
run_button.on_click(_render_results)

factor_select.options = _filtered_factor_options("")
factor_select.value = ""
_refresh_pinned_panel()
_render_results()

left_panel = widgets.VBox(
    [
        widgets.HTML("<h3 style='margin:0 0 8px 0;'>Factor Universe A</h3>"),
        search_box,
        factor_select,
    ],
    layout=widgets.Layout(width="42%", min_width="300px"),
)

right_panel = widgets.VBox(
    [
        widgets.HTML("<h3 style='margin:0 0 8px 0;'>Pinned Shocks</h3>"),
        pinned_box,
        widgets.HBox(
            [
                widgets.HTML("<b style='line-height:28px;'>Covariance</b>"),
                sample_checkbox,
                ewma_checkbox,
            ],
            layout=widgets.Layout(margin="8px 0 0 0", gap="10px", align_items="center"),
        ),
        widgets.HBox([run_button, clear_button], layout=widgets.Layout(margin="8px 0 0 0")),
    ],
    layout=widgets.Layout(width="58%", min_width="360px"),
)

dashboard = widgets.VBox(
    [
        widgets.HBox(
            [left_panel, right_panel],
            layout=widgets.Layout(width="100%", gap="18px", align_items="flex-start"),
        ),
        result_output,
    ],
    layout=widgets.Layout(width="100%"),
)

display(dashboard)
```
